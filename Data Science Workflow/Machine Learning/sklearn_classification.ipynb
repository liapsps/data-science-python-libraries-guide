# Classificação com Scikit-Learn
Neste exemplo, vamos treinar um modelo para classificar flores (dataset Iris clássico) em suas respectivas espécies.

**Modelo usado:** Árvore de Decisão (Decision Tree).

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# 1. Carregando dados
iris = load_iris()
X = iris.data
y = iris.target

# 2. Divisão Treino/Teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 3. Treino
clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)

# 4. Avaliação
previsoes = clf.predict(X_test)
acuracia = accuracy_score(y_test, previsoes)

print(f"Acurácia do modelo: {acuracia * 100:.2f}%")
```
