# Gráficos Interativos com Plotly
O **Plotly** permite criar gráficos interativos onde o usuário pode dar zoom, passar o mouse para ver detalhes (hover) e filtrar dados dinamicamente. Ideal para dashboards e apresentações web.

```python
import plotly.express as px
import pandas as pd

# Dados fictícios: Performance de funcionários
df = pd.DataFrame({
    'Funcionario': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Vendas': [1200, 1500, 900, 1700, 1300],
    'Satisfacao_Cliente': [4.5, 4.2, 3.8, 4.8, 4.0],
    'Regiao': ['Norte', 'Sul', 'Norte', 'Sul', 'Leste']
})

# Gráfico de Dispersão Interativo
fig = px.scatter(df, 
                 x='Vendas', 
                 y='Satisfacao_Cliente', 
                 color='Regiao',
                 size='Vendas', 
                 hover_name='Funcionario',
                 title='Vendas vs Satisfação do Cliente (Interativo)')

fig.show()
```
