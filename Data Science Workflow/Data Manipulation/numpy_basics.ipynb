# Introdução ao NumPy
O **NumPy** (Numerical Python) é a biblioteca fundamental para computação científica em Python. Ela fornece suporte para arrays multidimensionais (matrizes), além de uma coleção de funções matemáticas de alto nível para operar nesses arrays.

**Neste notebook, vamos aprender:**
1. Como criar arrays NumPy.
2. Indexação e fatiamento (slicing).
3. Operações matemáticas básicas.

```python
# Importando a biblioteca
import numpy as np

# A convenção padrão é importar o numpy como 'np'
print(f"Versão do NumPy: {np.__version__}")
```

```python
# 1. Criando Arrays
# Criando uma lista Python comum
minha_lista = [1, 2, 3, 4, 5]

# Convertendo para um array NumPy
meu_array = np.array(minha_lista)

print("Array criado:", meu_array)
print("Tipo do objeto:", type(meu_array))
```

```python
# 2. Arrays Multidimensionais (Matrizes)
# Criando uma matriz 2x3 (2 linhas, 3 colunas)
matriz = np.array([[1, 2, 3], [4, 5, 6]])

print("Matriz:\n", matriz)
print("Dimensões (shape):", matriz.shape)
```

```python
# 3. Operações Matemáticas
# NumPy permite operações elemento a elemento de forma muito rápida
array_a = np.array([10, 20, 30])
array_b = np.array([1, 2, 3])

soma = array_a + array_b
multiplicacao = array_a * 2

print("Soma dos arrays:", soma)
print("Multiplicação por escalar:", multiplicacao)

# Estatísticas básicas
print("Média dos valores:", array_a.mean())
print("Valor máximo:", array_a.max())
```
